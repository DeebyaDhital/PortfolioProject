```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #adjust the configuration of the plots we will create
```

```python
df=pd.read_csv(r'C:\Users\Laptop\Downloads\movies.csv')
```

```python
df.head()
```

```python
# lets see if there is any missing data
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df=df.dropna()
```

```python
#Data types for our columns
df.dtypes
```

```python
#changing datatypes of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scatter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs gross earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for films')
plt.show()
```

```python
#plot budget vs gross using seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#lets start looking at correlations
df.corr(method='pearson')#methods=pearson,spearman,kendall
```

```python
#lets start looking at correlations
#high correlation between budget and gross
correlation_matrix=df.corr(method='pearson')#methods=pearson,spearman,kendall
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
#lets look at company correlation now first lets convert the string vlaue into numeric
df.head()
```

```python
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix=df_numerized.corr(method='pearson')#methods=pearson,spearman,kendall
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
